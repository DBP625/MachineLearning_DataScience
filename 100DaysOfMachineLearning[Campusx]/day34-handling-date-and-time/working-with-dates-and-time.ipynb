# Working with Dates
### Extracting date from date

````python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
````

````python
date.info()
````

### 1. Extract year

````python
date['date_year'] = date['date'].dt.year
date.sample(5)
````

#### 2. Extract Month

````python
date['date_month_no'] = date['date'].dt.month
date.head()
````

### Though the month name is not here , We are mentioning it

````python
date['date_month_name'] = date['date'].dt.month_name()
date.head()
````

#### Extract Days

````python
# Extracting Days
date['date_day'] = date['date'].dt.day
date.head()
````

````python
# day of week
date['date_dow'] = date['date'].dt.dayofweek
date.head()
````

````python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()
date.head()

# date.drop(columns=['product_id','city_id','orders']).head()
````

````python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Friday' , 'Saturday']) , 1,0)


date.drop(columns=['product_id','city_id','orders']).head()


date.sample(100)
````

#### Extract week of the year

````python
date['date_week'] = date['date'].dt.isocalendar().week

date.drop(columns=['product_id','city_id','orders']).head()
````

#### Extract Quarter

1. Q1 (Quarter 1): January, February, March
2. Q2 (Quarter 2): April, May, June
3. Q3 (Quarter 3): July, August, September
4. Q4 (Quarter 4): October, November, December

````python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
````

#### Extract Semester

````python
date['semester'] = np.where(date['quarter'].isin([1,2]) , 1 , 2)
date.drop(columns=['product_id','city_id','orders']).head()
````

````python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
````
