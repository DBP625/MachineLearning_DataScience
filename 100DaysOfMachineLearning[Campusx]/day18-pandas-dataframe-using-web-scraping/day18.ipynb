```python

name=[]
processor=[]
ram=[]
storage=[]
display=[]
features=[]
price=[]

laptop_items = soup.find_all('div',class_='p-item')

for item in laptop_items:
    #Getting laptop name
    laptop_name = item.find('h4',class_='p-item-name').text.strip()
    name.append(laptop_name)
    len(name)
    
    laptop_price = item.find('div',class_='p-item-price').text.strip()
    price.append(laptop_price)
    len(price)
    
    #Get Specifications from short description
    specs = item.select('div.short-description li')
    len(specs)
    
    
    # Initialize flags to track if we found each spec
    found_processor = False
    found_ram = False
    found_storage = False
    found_display = False
    found_features = False
    found_price = False
    
    
    for spec in specs:
        text = spec.text.strip()
        if 'Processor' in text:
            processor.append(text.replace('Processor:','').strip())
            found_processor = True
            
        if 'RAM' in text or 'Ram' in text in spec.text:
             text = spec.text.replace(';',',') #replacing semicolon to comma
             parts = text.split(',')
             ram_part = parts[0].lower().replace('ram:','').strip()
             ram.append(ram_part)
             found_ram=True
             
             if len(parts)>1:
                 storage_part = parts[1].lower().replace('storage:','').strip()
                 storage.append(storage_part)
                 found_storage = True

        if 'Display' in text:
            display.append(text.replace('Display:','').strip())
            found_display = True
            
        if 'Features' in text:
            features.append(text.replace('Features:','').strip())
            found_features = True
        
    if not found_processor: processor.append(np.nan)
    if not found_ram: ram.append(np.nan)
    if not found_storage: storage.append(np.nan)
    if not found_display: display.append(np.nan)
    if not found_features: features.append(np.nan)
        
    
    d = {
    'Name': name,
    'Processor': processor,
    'RAM': ram,
    'Storage': storage,
    'Display': display,
    'Features': features,
    'Price': price
    }
    df=pd.DataFrame(d) 

    
    
    
    

        
            
```

# creating dataframe for all the pages

```python
final=pd.DataFrame()




for j in range(1,10):
  webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(j)).text
  soup=BeautifulSoup(webpage,'lxml')
  company=soup.find_all('div',class_='company-content-wrapper')
  name=[]
  rating=[]
  reviews=[]
  ctype=[]
  hq=[]
  how_old=[]
  no_of_employee=[]

  for i in company:

    try:
       name.append(i.find('h2').text.strip())
    except:
       name.append(np.nan)

    try:
       rating.append(i.find('p',class_='rating').text.strip())
    except:
       rating.append(np.nan)
   
    try:

      reviews.append(i.find('a' , class_='review-count').text.strip())
    except:
      reviews.append(np.nan)

    try:

      ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    except:
      ctype.append(np.nan)
    try:

      hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    except:
      hq.append(np.nan)
    
    try:

      how_old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    except:
      how_old.append(np.nan)
    try:
      no_of_employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    except:
      no_of_employee.append(np.nan)
    

  df=pd.DataFrame({'name':name,
    'rating':rating,
    'reviews':reviews,
    'company_type':ctype,
    'Head_Quarters':hq,
    'Company_Age':how_old,
    'No_of_Employee':no_of_employee,
    })
  
  final=final.append(df,ignore_index=True)



```
